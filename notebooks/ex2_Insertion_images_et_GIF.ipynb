## Enoncé et description de la méthode

Il n'est pas toujours possible pour une fonction arbitraire de trouver la forme explicite d'une primitive. L'idée principale est alors  de chercher des méthodes de calcul approchées visant à  simplifier  le calcul de l'intégrale en remplaçant la fonction $f$ par un polynôme d'interpolation $P$ de la fonction en certains points $x_k$ appartenant à l'intervalle $[a,b]$. La précision de la méthode dépend directement du choix des noeuds d'interpolatipon $x_k$ et du  degré du  polynôme d'interpolation. 

$$ \int_{a}^{b} f(x) \ dx  \approx \int_{a}^{b} P(x) \ dx $$


Le polynôme d'interpolation s'exprimant dans la base de Lagrange en fonction des valeur de $f(x_k)$, on obtient :

$$ 
\int_{a}^{b} P(x) \ dx =\sum_{k=1}^{n}p_k \ f(x_k)$$ 
où :
* $n$ : nombre de noeuds
* $p_k$ : pondérations
* $x_k$ : noeuds d'interpolations. 

Cette formule de quadrature est assimilable à une moyenne pondérée. Le choix de $n$, des pondérations $p_k$ et des noeuds $x_k$ dépendent de la méthode employée.

### Méthodes classiques
* **Méthode du point milieu (n = 1) :**   
Le principe est d'assimiler la région sous la courbe de la fonction $f$ à un **rectange** et d'en calculer l'aire.   
$n=1$ et $x_{1}=(a+b)/2$ et $p_1=(b-a)$,
$$
\int_{a}^{b} f(x) \ dx \approx  p_1 f(x_1) 
$$

$$
\Longleftrightarrow \boxed{\int_{a}^{b} f(x) \ dx \approx  (b-a) \ f\left(\frac{a+b}{2} \right)} \ \ (aire \  d'un \ rectangle)
$$   

Cette méthode est exacte pour les polynômes de degré 1. C'est une méthode d'ordre 1.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/bc/Mittelpunktsregel.svg/220px-Mittelpunktsregel.svg.png" width="200" height="200" />

* **Méthode des trapèzes (n = 2) :**   
Le principe est d'assimiler la région sous la courbe de la fonction $f$ à un **trapèze** et d'en calculer l'aire.   
$n=2$ et $x_{1}=a$ et $x_{2}=b$ et $p_1=p_2=(b-a)/2$,
$$
\int_{a}^{b} f(x) \ dx \approx p_1 f(x_1)+p_2 f(x_2)
$$

$$
\Longleftrightarrow \boxed{\int_{a}^{b} f(x) \ dx \approx (b-a) \frac{f(a)+f(b)}{2}}  \ \ (aire \  d'un \ trapeze)
$$

Cette méthode est exacte pour les polynômes de degré 1. C'est une méthode d'ordre 1.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Trapezoidal_rule_illustration.png/220px-Trapezoidal_rule_illustration.png" width="200" height="200" />

* **Méthode de Simpson (n = 3) :**   
Cette méthode utilise l'approximation de $f$ par un **polynôme $P$ de degré 2** de la forme $P(x)= \alpha x^2 + \beta x + \gamma $ qui interpole les 3 points $x_1=a$, $x_2=m= (a + b)/2$ et $x_3=b$. 
La formule de quadrature est la suivante :  
$n=3$ et $x_{1} = a$ et $x_{2}=(a+b)/2$ et $x_{3}=b$  et $p_1=p_3=(b-a)/6$ et $p_2=4(b-a)/6$ 
$$
\int_{a}^{b} f(x) \ dx \approx p_1 f(x_1)+p_2 f(x_2)+p_3 f(x_3)  
$$

$$
\Longleftrightarrow \boxed{\int_{a}^{b} f(x)dx \approx  \frac{b-a}{6}\left[f(a) + 4f\left(\frac{a+b}{2}\right)+f(b)\right]}. 
$$

Cette méthode est exacte pour les polynômes de degré 3. C'est une méthode d'ordre 3.

### Précision
Pour chacune des méthodes précédentes, le terme d'erreur dépend de $(b-a)$. Si cette amplitude est trop élevée, on peut réduire simplement l'erreur en découpant l'intervalle $[a, b]$ en **$M$ sous-intervalles**, sur lesquels on calculera la valeur approchée de l'intégrale. On parle alors de formule composite. La valeur sur l'intervalle $[a, b]$ sera la **somme** de la valeur sur chaque sous-intervalle :

$$
\int_{a}^{b} f(x) \ dx =\sum_{k=1}^{M}\int_{x_{k-1}}^{x_{k}} f(x) \ dx
$$
